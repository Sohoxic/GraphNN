```python
high_degree_nodes = [node for node in G.nodes if G.degree[node] > 10]
len(list(high_degree_nodes))
```

```python
def pagerank_matrix(A, alpha=0.85):
    N = A.shape[0]
    D = np.sum(A, axis=1)
    P = (A.T / D).T
    P = alpha * P + (1 - alpha) * np.ones((N, N)) / N
    return P

def personalized_pagerank(P, u, alpha=0.85, epsilon=1e-8):
    N = P.shape[0]
    r = np.zeros(N)
    r[u]
```

```python
H = G.subgraph(high_degree_nodes)
```

## Inference: The nodes are more likely to jump to  another  page rather than to follow a link which could be understood from the value of alpha in page rank

```python
pr = nx.pagerank(H, alpha=0.1)
pr
```

```python
communities = nx.algorithms.community.girvan_newman(H)
```

```python
for c in next(communities):
  print(c)
```

```python
import itertools
#first k tuples of communities
k=2
for comm in itertools.islice(communities, k):
    print(tuple(sorted(c) for c in comm))
```

```python
# compute eigenvector centrality
eigenvector_centrality = nx.algorithms.eigenvector_centrality(H)

# print nodes with highest eigenvector centrality
for node, eigenvector in sorted(eigenvector_centrality.items(), key=lambda x: x[1], reverse=True):
    print(node, eigenvector)
```

## `Top 10` most influencial nodes:

```python
# print the top 10 nodes with highest eigenvector centrality
print("Top 10 nodes with highest eigenvector centrality:")
for i, (node, eigenvector) in enumerate(sorted(eigenvector_centrality.items(), key=lambda x: x[1], reverse=True)):
    print(f"{i+1}. Node {node}: {eigenvector}")
    if i == 9:
        break
```

## Betweeness centrality is `less efficient` for this dataset as it involves going through each and every shortest path reachable from the given node to a particular node divided by the shortest path reachable from a given node to a particular node via another node. It is the summation of that




```python
betweenness_centrality = nx.algorithms.betweenness_centrality(H)

# print nodes with highest betweenness centrality
for node, betweenness in sorted(betweenness_centrality.items(), key=lambda x: x[1], reverse=True):
    print(node, betweenness)
```
